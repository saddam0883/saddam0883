```python
import scipy.stats as stats
from scipy.stats import ttest_1samp, ttest_ind, wilcoxon
from statsmodels.stats.power import ttest_power
```

```python
smoker = df[df['smoker']=='yes']['charges']
nonsmoker= df[df['smoker']=='no']['charges']
sns.distplot(smoker)
sns.distplot(nonsmoker)
```

Statistically determine whether Charges for Smokers and Nonsmokers are significantly different or not.
Charges is Numerical variable and Smoker is Categorical variable.
Smokers and Nonsmokers are two independent samples(a smoker cannot be a nonsmoker at the same time).
Using two sample t-test

H0 : Charges of smoker doesn't differ significantly from charges of non smoker.
HA : Charges of smoker differ significantly from charges of non-smoker.
Here we fix level of significance as  Î±  = 0.05.

Alternative hypothesis starts from the view of what statistician is going to prove, here we are trying to prove Charges of smoker differ significantly from charges of non-smoker.

```python
stats.ttest_ind(smoker,nonsmoker)
```

p-value is less than level of significance, hence reject Null hypothesis, which means alternative hypothesis is true, hence Charges of smoker differ significantly from charges of non-smoker.

```python
male = df[df['sex']=='male']['bmi']
female = df[df['sex']=='female']['bmi']
sns.distplot(male)
sns.distplot(female)
```

Statistically determine whether BMI of Males and Females are significantly different or not.
BMI is Numerical variable and Sex is Categorical variable.
Using two sample t-test.

H0 : BMI of Male and Female are not significantly different.
HA : BMI of Male and Female are significantly different.
level of significance as  Î±  = 0.05.

```python
stats.ttest_ind(male,female)
```

p-value is greater than level of significance, hence do not reject Null hypothesis, hence BMI of Male is not so different from Female.

```python
from scipy.stats import f_oneway
```

Using Chi_square test to check if Smoker proportion is significanty different for different Genders.
Smoker and Sex are Categorical variables

H0 : The proportion of smokers aren't significantly different in different genders.
HA : The proportion of smokers significantly different in different genders.

```python
chi2,pval,dof,expected = stats.chi2_contingency(crosst)
pval
```
